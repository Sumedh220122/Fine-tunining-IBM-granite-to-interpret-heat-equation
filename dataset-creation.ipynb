```python
def generate_vtk(solution_func, filename, nx=100, ny=100):
    # Create mesh grid
    x = np.linspace(0, 1, nx)
    y = np.linspace(0, 1, ny)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X) 
    
    temperature = solution_func(X, Y)
    
    grid = pv.StructuredGrid(X, Y, Z)
    grid["Temperature"] = temperature.ravel(order="F") 
    
    grid.save("./working/" + filename + ".vtk")
```

```python
def load_and_plot_vtk(filename):
    mesh = pv.read(filename)
    # Check if the file has the expected data
    print("Available Scalars:", mesh.array_names)
    
    # Plot the temperature field
    if "Temperature" in mesh.array_names:
        points = mesh.points  # (x, y, z) coordinates
        temperature = mesh['Temperature']
        print(points.shape)
        x = points[:, 0].reshape(100, 100)
        y = points[:, 1].reshape(100, 100)
        temperature = temperature.reshape(100, 100)
        plt.contourf(x, y, temperature, cmap='jet', levels=100)
        plt.colorbar()
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Temperature')
        plt.show()
    else:
        print("No 'Temperature' data found in the file.")
```

```python
def generate_analysis_qa_pairs(): # Function to generate data for case 2q1
    qa_pairs = []
    y = 0.5 
    
    for x in np.linspace(0, 1, 200): 
        temperature = x**2 + y**2 
        
        question = f"Analyze this steady-state heat equation solution T(x, y) = x^2 + y^2 in a unit square domain and tell how does the temperature change with respect to the position along the x-axis at y = 0.5?"
        
        answer = f"As we move along the x-axis at y = 0.5, the temperature increases quadratically with x. Specifically, T(x, 0.5) = x^2 + 0.25. For example:\n" \
                 f"At x = {x:.1f}, the temperature is {temperature:.2f}."
        
        qa_pairs.append((question, answer))
    
    return qa_pairs
```

```python
def generate_qa_pairs_for_increased_pi():
    qa_pairs = []
    
    y = 0.5  # Fixed y-value
    
    x_values = np.linspace(0, 1, 200)
    
    for x in x_values:
        temperature = x**2 + y**2 
        
        # Construct the question
        question = "Analyze this steady-state heat equation solution T(x, y) = x^2 + y^2 in a unit square domain and tell if we increase the coefficient of pi in the force function, what will happen?"
        
        answer = f"By increasing the coefficient of pi in the force function, the temperature will generally increase across the domain. The increased forcing term would result in higher temperature values, including at the point (x = {x:.1f}, y = 0.5). For example, the temperature at x = {x:.1f} and y = 0.5 is {temperature:.2f} in the baseline case."
        
        qa_pairs.append((question, answer))
    
    return qa_pairs
```

```python
def generate_qa_for_zero_temperature_at_boundaries():
    qa_pairs = []

    def temperature(x, y):
        return x**2 + y**2

    y = 0.5
    x_values = np.linspace(0, 1, 200)

    for x in x_values:
        T = temperature(x, y)
        
        question = """
        Analyze this steady-state heat equation solution T(x, y) = x^2 + y^2 in a unit square domain 
        and explain why the temperature is zero at both x=0 and x=1, and what this means physically.
        """
        
        answer = f"""
        The temperature solution T(x, y) = x^2 + y^2 is not inherently zero at both x=0 and x=1 for all y.
        However, boundary conditions may be applied that enforce T(0, y) = 0 and T(1, y) = 0 for all y.
        This would imply that the temperature is fixed at zero at the boundaries.

        At x = {x:.2f}, the temperature is {T:.2f}, which is the result of the equation T(x, y) = x^2 + y^2.
        Physically, this might represent a system with insulated or fixed temperature boundaries at x=0 and x=1.
        """
        
        qa_pairs.append((question, answer))

    return qa_pairs
```

```python
def generate_case2q2_for_100_values():
    qa_pairs = []
    y = 0.5
    x_values = np.linspace(0, 1, 200)
    
    for x in x_values:
        question = """
        Analyze this steady-state heat equation solution T(x, y) = x^2 + y^2 in a unit square domain and tell 
        at what coordinates does the maximum temperature occur, and what determines this location?
        """
        
        answer = f"""
        For x = {x:.2f}, the temperature is T({x:.2f}, {y}) = {x**2 + y**2:.2f}. The temperature function T(x, y) = x^2 + y^2 is maximized when both x and y are at their maximum values. 
        In the unit square domain, the maximum temperature occurs at the point (1, 1), determined by the fact that both x^2 and y^2 increase as x and y approach 1.
        """
        
        qa_pairs.append((question, answer))
    
    return qa_pairs
```

```python
def generate_case2q3_for_100_values():
    qa_pairs = []
    y_values = np.linspace(0, 1, 200)
    
    for y in y_values:
        question = """
        Analyze this steady-state heat equation solution T(x, y) = x^2 + y^2 in a unit square domain and tell 
        how does the temperature profile change along the vertical line x=0.5 compared to x=0.25?
        """
        
        answer = f"""
        Along the vertical line x=0.5, the temperature at y = {y:.2f} is T(0.5, {y:.2f}) = {0.5**2 + y**2:.2f}.
        Similarly, along the vertical line x=0.25, the temperature at y = {y:.2f} is T(0.25, {y:.2f}) = {0.25**2 + y**2:.2f}.
        As x increases, the temperature increases because the x^2 term increases. Therefore, the temperature along the line x=0.5 will be 
        higher than along the line x=0.25 for the same values of y.
        """
        
        qa_pairs.append((question, answer))
    
    return qa_pairs

def generate_case3q1_for_100_values():
    qa_pairs = []
    x_values = np.linspace(0, 1, 200)
    
    for x in x_values:
        question = """
        Analyze this steady-state heat equation solution T(x, y) = x^2 + y^2 in a unit square domain and tell 
        what is the temperature at the corner (0, 0) of the unit square mesh?
        """
        
        answer = f"""
        At the corner (0, 0), the temperature is always T(0, 0) = 0^2 + 0^2 = 0, regardless of x.
        This is the temperature at the bottom-left corner of the unit square.
        """
        
        qa_pairs.append((question, answer))
    
    return qa_pairs

def generate_case3q2_for_100_values():
    qa_pairs = []
    y_values = np.linspace(0, 1, 200)
    
    for y in y_values:
        question = """
        Analyze this steady-state heat equation solution T(x, y) = x^2 + y^2 in a unit square domain and tell 
        what physical significance does the boundary condition u(0, y) = 0 have in the context of heat diffusion on the unit square mesh?
        """
        
        answer = f"""
        The boundary condition u(0, {y:.2f}) = 0 indicates that the temperature is fixed to zero along the left boundary 
        (x=0) of the unit square. This suggests that there is no heat being applied or flowing into the domain from 
        the left side, and that side is insulated or kept at a fixed temperature. This is a typical boundary condition 
        for heat diffusion problems, indicating no heat source or a thermal reservoir at the boundary.
        """
        
        qa_pairs.append((question, answer))
    
    return qa_pairs
```

```python
func_list =[generate_analysis_qa_pairs, generate_qa_pairs_for_increased_pi,
            generate_qa_for_zero_temperature_at_boundaries,
            generate_case2q2_for_100_values,
            generate_case2q3_for_100_values, 
            generate_case3q1_for_100_values,
            generate_case3q2_for_100_values]
```

```python
def append_qa_pairs_to_df(qa_pairs, df):
    qa_df = pd.DataFrame(qa_pairs, columns=["Questions", "Answers"])
    df = pd.concat([df, qa_df], ignore_index=True)
    return df
```

```python
import pandas as pd

df = pd.DataFrame(columns = ["Questions", "Answers"])

for fun in func_list:
    qa_pairs = fun()
    df = append_qa_pairs_to_df(qa_pairs, df)
```
